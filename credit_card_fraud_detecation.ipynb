**importing necessary libraries**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
cd=pd.read_csv('/content/drive/MyDrive/creditcard.csv')  # csv=comma seprated values
cd.head(6)
```

**this dataset is highly unbalanced**

**0 indicates noramal transacton**

**1 indicaates fraudulent transaction**

```python
# distribution of legit transacton and frauduleant transection

cd['Class'].value_counts()
```

```python

# seprating the data for analysis

legit=cd[cd.Class==0]
fraud=cd[cd.Class==1]

print(legit.shape)
print()
print(fraud.shape)
```

```python
# statistical measure of data

legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the value for both transactions

cd.groupby('Class').mean()
```

**Under Sampling**

**Build the dataset containing similar distribution of normal transactions and fraudulent transacton**

**number of fraudulent transaction=492**

```python
legit_sample=legit.sample(n=492)
```

**Concatenating two dataframes**

```python
new_dataset=pd.concat([legit_sample, fraud],axis=0)
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

**spliting data into target and features**

```python
x=new_dataset.drop(columns='Class',axis=1)
print(x)
```

```python
y=new_dataset['Class']
print(y)
```

**Spliting Data into Training data and Testing data**

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2,stratify=y, random_state=2)
print(x.shape, x_train.shape, x_test.shape)
```

**Model Training **

**LOGESTIC REGRESSION**

```python
model=LogisticRegression()
```

```python
# training the logistic regression model with training data

model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)*100
```

**MODEL EVALUTION**

**ACCURACY SCORE**

```python
# accuracy on training data

x_train_prediction=model.predict(x_train)

training_data_accuracy=accuracy_score(x_train_prediction, y_train)*100

print('accuracy on training data:',training_data_accuracy)
```

```python
# accuracy on test data

x_test_prediction=model.predict(x_test)

testing_data_accuracy=accuracy_score(x_test_prediction, y_test)*100

print('accuracy on testing data:',testing_data_accuracy)
```
